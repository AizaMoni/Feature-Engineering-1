```python
#Loop
columns_to_normalize = ['Marketing Spend', 'Administration', 'Transport']
# Loop through each column and normalize
for column in columns_to_normalize:
    x[[column]] = m.fit_transform(x[[column]])

# Print the normalized data
print(x)
```
